```python
import matplotlib.pyplot as plt

# Data extracted from the image
frameworks = ['Manual', 'H2O', 'AutoKeras', 'AutoGluon']
scores = [0.9834, 0.9711, 0.8615, 0.9795]

# Plotting the bar graph
plt.figure(figsize=(12, 8))
plt.bar(frameworks, scores, color=['blue', 'green', 'red', 'purple'])

# Annotating each bar with the corresponding value
for i, score in enumerate(scores):
    plt.text(i, score + 0.005, f'{score:.4f}', ha='center', va='bottom', fontsize=12)

# Adding labels and title
plt.xlabel('Frameworks', fontsize = 20)
plt.ylabel('Performance Score', fontsize = 20)
plt.title('Performance Scores by Framework',fontsize = 20)
plt.ylim(0.85, 1.0)  # Set y-axis limits for better visualization
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid(axis='y')

# Show the plot
# plt.show()
plt.savefig('./MNIST_automl.jpeg')
```
