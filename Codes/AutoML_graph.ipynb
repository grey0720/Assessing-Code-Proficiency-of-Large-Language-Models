```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터
datasets = ["Titanic", "MNIST", "Steel Defect"]
h2o_scores = [0.732, 0.9795, 0.8815]
autokeras_scores = [0.7441, 0.8615, 0.8565]
autogluon_scores = [0.7631, 0.9711, 0.8837]
best_scores = [0.7909, 0.9834, 0.8738]

# 막대 너비 및 위치 설정
x = np.arange(len(datasets))
bar_width = 0.2

# 그래프 크기 설정
plt.figure(figsize=(12, 8))

# 막대 그래프 그리기
plt.bar(x - bar_width, h2o_scores, width=bar_width, label="H2O", color="blue")
plt.bar(x, autokeras_scores, width=bar_width, label="AutoKeras", color="green")
plt.bar(x + bar_width, autogluon_scores, width=bar_width, label="AutoGluon", color="orange")

# 최고 점수 표시
for i in range(len(datasets)):
    plt.scatter(i, best_scores[i], color="red", zorder=5, label="Average Score of Dev." if i == 0 else "")

# 최고 점수 연결하는 선 추가
plt.plot(x, best_scores, color="red", linestyle="--", marker="o", label="_nolegend_")

# 각 막대 위에 점수 표시
for i in range(len(datasets)):
    plt.text(i - bar_width, h2o_scores[i] + 0.001, f"{h2o_scores[i]:.4f}", ha="center", fontsize=12)
    plt.text(i, autokeras_scores[i] + 0.001, f"{autokeras_scores[i]:.4f}", ha="center", fontsize=12)
    plt.text(i + bar_width, autogluon_scores[i] + 0.001, f"{autogluon_scores[i]:.4f}", ha="center", fontsize=12)

    if i == 0:  # Titanic에 해당하는 최고 점수 0.7966을 아래로 표시
        plt.text(i, best_scores[i] - 0.013, f"{best_scores[i]:.4f}", ha="center", color="red", fontsize=12)
    else:  # 다른 값들은 위에 표시
        plt.text(i, best_scores[i] + 0.006, f"{best_scores[i]:.4f}", ha="center", color="red", fontsize=12)

# 그래프 제목과 레이블 설정
plt.title("Auto ML Results and Best Score of Dev.", fontsize=30)
plt.xlabel("Dataset", fontsize=25)
plt.ylabel("Score", fontsize=25)
plt.xticks(x, datasets, fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(0.7, 1.0)
plt.grid(axis = 'y')

# 범례 추가
plt.legend(loc="upper left", fontsize=15)

# 그래프 표시
plt.tight_layout()
# plt.show()
plt.savefig('./automl_bestscore.jpeg')
plt.savefig('./automl_bestscore.pdf')
```
