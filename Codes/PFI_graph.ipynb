```python
import matplotlib.pyplot as plt
import numpy as np

# PFI values for each classifier
pfi_data = {
    "Random Forest": [0.094, 0.067, 0.068, 0.049, None],
    "XGBoost": [0.055, 0.107, 0.075, 0.030, None],
    "Voting Classifier": [0.071, 0.105, 0.048, 0.029, None]
}

# Features to plot
features = ["Sex", "Title", "Pclass", "Fare", "IsAlone"]
colors = {"Random Forest": "red", "XGBoost": "blue", "Voting Classifier": "green"}

x = np.arange(len(features))
bar_width = 0.2

plt.figure(figsize=(16, 12))

# Plotting each classifier's PFI values
for idx, (classifier, color) in enumerate(colors.items()):
    pfi_values = [val if val is not None else 0 for val in pfi_data[classifier]]
    bars = plt.bar(x + idx * bar_width, pfi_values, width=bar_width, label=classifier, color=color)

    # Adding value labels
    for bar, val in zip(bars, pfi_values):
        if val > 0:  # Only label non-zero values
            plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                     f'{val:.3f}', ha='center', va='bottom', fontsize=13)

# Labeling and setting y-axis range
plt.xlabel("Features", fontsize = 25)
plt.ylabel("Average PFI Value", fontsize = 25)
plt.title("Average PFI Values for Different Classifiers - Developer", fontsize = 30)
plt.xticks(x + bar_width, features, fontsize = 30)
plt.yticks(fontsize = 25)
plt.ylim(0, 0.2)  # Setting y-axis range
plt.grid(axis = 'y')
plt.legend(fontsize = 25)
# plt.show()

plt.savefig('./pfi-dev.pdf')
plt.savefig('./pfi-dev.jpeg')
```
